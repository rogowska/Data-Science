## Exercise 2
Using scatter plot:  
a) Show the relation of square footage and price.  
b) Check how price is influenced by apartment's grade. Find other variables correlated with price.  
c) Check how good apartments are distributed over the city. First, plot zipcode versus price. Then, use apartments coordinates (longitude and lattitude) in order to show where apartments of good grade are located.

```python
#a)
#Your code:
plt.figure(figsize = (15,10))
plt.scatter(df['sqft_living'], df['price'])
plt.xlabel("Square footage")
plt.ylabel("Price")
plt.show()

#We can notice a relation between square footage and price - the bigger square footage is, the price is higher. This relation is nearly linear.
```

```python
#b)
#Your code:
plt.figure(figsize = (15,10))
plt.scatter(df['grade'], df['price'])
plt.xlabel("Grade")
plt.ylabel("Price")
plt.show()

#We can notice that more expensive houses are better graded, and the ones poorly graded are mostly cheap.
```

```python
#c)
#Your code:
plt.figure(figsize = (15,10))
plt.scatter(df['zipcode'], df['price'])
plt.xlabel("Zipcode")
plt.ylabel("Price")
plt.show()

#We can notice that price isnt noticeably corelated with the zipcode.
```

```python
#Your code:
plt.figure(figsize = (15,10))
plt.scatter(df.long, df.lat, c=df.grade, cmap='viridis')
plt.colorbar(label = 'grade')
plt.xlabel("Longitude")
plt.ylabel("Lattitude")
plt.show()

#We can notice that the best graded houses are in the centre of the map.
```

## Exercise 4
"I am interested in the house with 3 bedrooms, 2 bathrooms and 2 floors"  
a) Compute the basic statistics of apartments fulfilling above constraints.  
b) Distinct accepted apartments on any of previously created plots.  
c) Find such apartments with best price to footage ratio.  

```python
#a)
#Your code:
chosen = df.loc[(df['bedrooms'] == 3) & (df['bathrooms'] == 2) & (df['floors'] == 2)]
chosen.describe()

#Hint: pandas.DataFrame.loc
```

```python
#b)
#Your code:
plt.figure(figsize = (15,10))
plt.scatter(df['sqft_living'], df['price'])
plt.scatter(chosen['sqft_living'], chosen['price'])
plt.xlabel("Square footage")
plt.ylabel("Price")
plt.show()
```

```python
#c)
#Your code:
chosen['best_ratio'] = chosen['price'] / df['sqft_living']
chosen.sort_values('best_ratio', ascending = True)

#Hint: .loc, .sort_values(), .index
```
